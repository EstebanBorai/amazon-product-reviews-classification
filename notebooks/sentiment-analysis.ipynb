```python
# NaN ratings are not useful, so these should be removed
df.dropna(subset=['rating'])
```

Our model should only classify english comments so predictions are more accurate. Reviews in other languages should be filtered out.

```python
from langdetect import detect

def detect_english(title, text):
    if detect(title) == 'en' and detect(text) == 'en':
        return 'en'
    return 'unknown'

df['lang'] = df.apply(lambda x: detect_english(x.title, x.text), axis=1)
```

```python
# Suspect theres some NaNs affecting `detect`
df.isna().sum()
```

Lets ease a bit the detection to allow either title or text to be English in order to determine wether the language is English, this could affect precision but might also bring in more samples to train the classification model.

```python
from langdetect import detect

def detect_english(title, text):
    if detect(title) == 'en' or detect(text) == 'en':
        return 'en'
    return 'unknown'

df['lang'] = df.apply(lambda x: detect_english(x.title, x.text), axis=1)
```

```python
df['lang'].value_counts()
```

This change brough in almost 98% of our samples! Now other items which are not English could be removed.
